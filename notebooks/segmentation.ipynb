```python
def map_mcc_to_category(mcc):
    # Handle missing or invalid MCC
    if pd.isna(mcc):
        return "Other"

    mcc = int(mcc)

    # RANGES FIRST
    if 3000 <= mcc <= 3350:
        return "Travel_Lodging"
    if 3500 <= mcc <= 3999:
        return "Travel_Lodging"

    # SINGLE MCC MAPS
    if mcc in {5411, 5499}:
        return "Groceries"
    if mcc in {5812, 5814}:
        return "Dining_Food"
    if mcc in {5541, 4411, 4121}:
        return "Transport_Fuel"
    if mcc in {5912}:
        return "Healthcare_Wellness"
    if mcc in {5651, 5733, 5311}:
        return "Retail_Apparel"
    if mcc in {4900}:
        return "Utilities_Bills"
    if mcc in {3075, 3144}:
        return "Travel_Lodging"
    if mcc in {4111, 4131, 4789}:
        return "Transport_Fuel"


# Retail expansion
    if mcc in {5300, 5399, 5699, 5732}:
        return "Retail_Apparel"


# Digital & subscriptions
    if mcc in {4814, 4899, 5815}:
        return "Digital_Subscriptions"


# Entertainment
    if mcc in {7832, 7996, 7997}:
        return "Entertainment_Leisure"

    return "Other"
```

```python
df3['spending_category'] = df3['MCC'].apply(map_mcc_to_category)

df3['spending_category'].value_counts(normalize=True)
```

```python
# User base (1 row per user)
user_base = df.copy()
user_base['User'] = user_base.index

# Sanity check
assert user_base['User'].is_unique
assert user_base.shape[0] == df.shape[0]
```

```python
# Convert 'Credit Limit' to numeric, removing '$' if present
df1['Credit Limit'] = df1['Credit Limit'].replace('[\$,]', '', regex=True).astype(float)

card_summary = (
    df1
    .groupby('User')
    .agg(
        num_cards=('CARD INDEX', 'count'),
        total_credit_limit=('Credit Limit', 'sum'),
        avg_credit_limit=('Credit Limit', 'mean'),
        chip_card_ratio=('Has Chip', lambda x: (x == 'YES').mean()),
        darkweb_card_ratio=('Card on Dark Web', lambda x: (x == 'Yes').mean())
    )
    .reset_index()
)
```

```python
user_base = user_base.merge(
    card_summary,
    on='User',
    how='left'
)

# Fill safe defaults for users with no cards
user_base[['num_cards', 'total_credit_limit', 'avg_credit_limit']] = (
    user_base[['num_cards', 'total_credit_limit', 'avg_credit_limit']].fillna(0)
)

user_base[['chip_card_ratio', 'darkweb_card_ratio']] = (
    user_base[['chip_card_ratio', 'darkweb_card_ratio']].fillna(0)
)

# Sanity check
assert user_base.shape[0] == df.shape[0]
```

```python
# Convert Amount column to numeric (remove $ and convert to float)
txn_base['Amount'] = txn_base['Amount'].replace('[\$,]', '', regex=True).astype(float)

txn_agg_basic = (
    txn_base
    .groupby('User')
    .agg(
        txn_count=('Amount', 'count'),
        total_spend=('Amount', 'sum'),
        avg_txn_amount=('Amount', 'mean'),
        spend_std=('Amount', 'std'),
        active_months=('Month', 'nunique')
    )
    .reset_index()
)

txn_agg_basic['spend_cv'] = (
    txn_agg_basic['spend_std'] / txn_agg_basic['avg_txn_amount']
).fillna(0)
```

```python
# Spend per category
cat_spend = (
    txn_base
    .pivot_table(
        index='User',
        columns='spending_category',
        values='Amount',
        aggfunc='sum',
        fill_value=0
    )
)

# Transaction count per category
cat_txn = (
    txn_base
    .pivot_table(
        index='User',
        columns='spending_category',
        values='Amount',
        aggfunc='count',
        fill_value=0
    )
)
```

```python
cat_spend.columns = [f"spend_{c}" for c in cat_spend.columns]
cat_txn.columns   = [f"txn_{c}" for c in cat_txn.columns]
```

```python
cat_spend_share = cat_spend.div(cat_spend.sum(axis=1), axis=0).fillna(0)
cat_spend_share.columns = [c.replace("spend_", "share_") for c in cat_spend.columns]

dominant_category = cat_spend_share.idxmax(axis=1)
```

```python
from scipy.stats import entropy

category_entropy = cat_spend_share.apply(
    lambda x: entropy(x[x > 0]),
    axis=1
)

category_entropy = category_entropy.rename("category_entropy")
```

```python
fraud_agg = (
    fraud_base
    .groupby('User')
    .agg(
        avg_fraud_score=('fraud_score', 'mean'),
        max_fraud_score=('fraud_score', 'max'),
        risk_volatility=('fraud_score', 'std'),
        pct_low=('fraud_band', lambda x: (x == 'Low').mean()),
        pct_medium=('fraud_band', lambda x: (x == 'Medium').mean()),
        pct_high=('fraud_band', lambda x: (x == 'High').mean()),
        pct_critical=('fraud_band', lambda x: (x == 'Critical').mean()),
        pct_decline=('fraud_action', lambda x: (x == 'Decline').mean())
    )
    .reset_index()
)

fraud_agg = fraud_agg.fillna(0)
```

```python
user_features = (
    user_base
    .merge(txn_agg_basic, on='User', how='left')
    .merge(cat_spend, left_on='User', right_index=True, how='left')
    .merge(cat_txn, left_on='User', right_index=True, how='left')
    .merge(cat_spend_share, left_on='User', right_index=True, how='left')
    .merge(dominant_category.rename("dominant_category"),
           left_on='User', right_index=True, how='left')
    .merge(category_entropy, left_on='User', right_index=True, how='left')
    .merge(fraud_agg, on='User', how='left')
)

user_features = user_features.fillna(0)
```

```python
# ---- feature lists ----
primary_feats = [
    'txn_count', 'total_spend', 'avg_txn_amount', 'spend_cv', 'category_entropy',
    'share_Groceries', 'share_Dining_Food', 'share_Transport_Fuel',
    'share_Retail_Apparel', 'share_Healthcare_Wellness',
    'share_Utilities_Bills', 'share_Digital_Subscriptions',
    'share_Travel_Lodging', 'share_Entertainment_Leisure'
]

secondary_feats = ['num_cards', 'total_credit_limit', 'avg_credit_limit']

tertiary_feats = ['avg_fraud_score', 'risk_volatility', 'pct_high', 'pct_critical']

cluster_feats = primary_feats + secondary_feats + tertiary_feats

X = user_features[cluster_feats].copy()
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler
from sklearn.compose import ColumnTransformer

robust_cols = ['txn_count', 'total_spend', 'avg_txn_amount',
               'total_credit_limit', 'avg_credit_limit']

standard_cols = [c for c in cluster_feats if c not in robust_cols]

preprocessor = ColumnTransformer(
    transformers=[
        ('robust', RobustScaler(), robust_cols),
        ('standard', StandardScaler(), standard_cols)
    ]
)

X_scaled = preprocessor.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.85, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

ks = range(3, 9)
sil_scores = []

for k in ks:
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    labels = km.fit_predict(X_pca)
    sil_scores.append(silhouette_score(X_pca, labels))

list(zip(ks, sil_scores))
```

```python
k_opt = 5  # update if silhouette suggests otherwise

kmeans = KMeans(n_clusters=k_opt, n_init=50, random_state=42)
cluster_labels = kmeans.fit_predict(X_pca)

user_features['cluster'] = cluster_labels
```

```python
# Refit with a different seed
kmeans_alt = KMeans(n_clusters=k_opt, n_init=50, random_state=7)
labels_alt = kmeans_alt.fit_predict(X_pca)

stability = (cluster_labels == labels_alt).mean()
stability
```

```python
import numpy as np
import pandas as pd

X = user_features[cluster_feats].copy()

# After log transform
heavy_cols = [
    'txn_count', 'total_spend', 'avg_txn_amount',
    'total_credit_limit', 'avg_credit_limit'
]

for col in heavy_cols:
    X[col] = np.log1p(X[col])

# Check invalids
invalid_summary = pd.DataFrame({
    'nan_count': X.isna().sum(),
    'inf_count': np.isinf(X).sum(),
    'zero_var': X.std() == 0
})

invalid_summary[invalid_summary.any(axis=1)]
```

```python
# Winsorize at 1–99%
for col in X.columns:
    lo, hi = X[col].quantile([0.01, 0.99])
    X[col] = X[col].clip(lo, hi)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import numpy as np
assert np.isfinite(X_scaled).all()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

ks = range(3, 9)
sil_scores = []

for k in ks:
    km = KMeans(n_clusters=k, n_init=30, random_state=42)
    labels = km.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels))

list(zip(ks, sil_scores))
```

```python
k_opt = 5  # adjust if silhouette clearly peaks elsewhere

km1 = KMeans(n_clusters=k_opt, n_init=50, random_state=42)
km2 = KMeans(n_clusters=k_opt, n_init=50, random_state=7)

labels_1 = km1.fit_predict(X_scaled)
labels_2 = km2.fit_predict(X_scaled)

stability = (labels_1 == labels_2).mean()
stability
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import numpy as np

ks = range(3, 8)
results = []

for k in ks:
    gmm = GaussianMixture(
        n_components=k,
        covariance_type='full',
        n_init=5,
        random_state=42
    )
    
    labels = gmm.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    bic = gmm.bic(X_scaled)
    
    results.append((k, sil, bic))

results
```

```python
k_opt = 5  # adjust based on BIC + silhouette

gmm = GaussianMixture(
    n_components=k_opt,
    covariance_type='full',
    n_init=10,
    random_state=42
)

cluster_labels = gmm.fit_predict(X_scaled)
cluster_probs = gmm.predict_proba(X_scaled)

user_features['cluster'] = cluster_labels
```

```python
gmm_alt = GaussianMixture(
    n_components=k_opt,
    covariance_type='full',
    n_init=10,
    random_state=7
)

labels_alt = gmm_alt.fit_predict(X_scaled)

stability = (cluster_labels == labels_alt).mean()
stability
```
