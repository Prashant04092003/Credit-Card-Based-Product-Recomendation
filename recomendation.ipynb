```python
def get_eligible_cards_for_user(user_row, cards_df):
    """
    Hard eligibility filter based on FICO.
    """
    fico = user_row['fico_score']

    eligible = cards_df[
        (cards_df['min_fico'] <= fico) &
        (cards_df['max_fico'] >= fico)
    ]

    return eligible
```

```python
def get_adjacent_clusters(cluster_id, max_cluster=6):
    """
    Returns adjacent cluster IDs (±1).
    """
    adjacent = []
    if cluster_id - 1 >= 0:
        adjacent.append(cluster_id - 1)
    if cluster_id + 1 <= max_cluster:
        adjacent.append(cluster_id + 1)
    return adjacent
```

```python
import numpy as np

def sample_cards_by_cluster(user_row, eligible_cards, total_cards=25):
    user_cluster = user_row['cluster']

    same_cluster = eligible_cards[
        eligible_cards['cluster_id'] == user_cluster
    ]

    adjacent_clusters = get_adjacent_clusters(user_cluster)
    adjacent = eligible_cards[
        eligible_cards['cluster_id'].isin(adjacent_clusters)
    ]

    other = eligible_cards[
        ~eligible_cards['cluster_id'].isin([user_cluster] + adjacent_clusters)
    ]

    n_same = int(0.6 * total_cards)
    n_adj  = int(0.25 * total_cards)
    n_other = total_cards - n_same - n_adj

    sampled = []

    if len(same_cluster) > 0:
        sampled.append(same_cluster.sample(
            min(n_same, len(same_cluster)), random_state=42))

    if len(adjacent) > 0:
        sampled.append(adjacent.sample(
            min(n_adj, len(adjacent)), random_state=42))

    if len(other) > 0:
        sampled.append(other.sample(
            min(n_other, len(other)), random_state=42))

    sampled_df = pd.concat(sampled).drop_duplicates()

    return sampled_df
```

```python
def enforce_tier_diversity(candidates, cards_df, min_per_tier=1):
    tiers = candidates['card_tier'].unique().tolist()

    required_tiers = ['starter', 'cashback', 'lifestyle']

    for tier in required_tiers:
        if tier not in tiers:
            extra = cards_df[cards_df['card_tier'] == tier]
            if len(extra) > 0:
                candidates = pd.concat([
                    candidates,
                    extra.sample(1, random_state=42)
                ])

    return candidates.drop_duplicates()
```

```python
def generate_candidate_cards(user_row, cards_df, total_cards=25):
    # Step 1: eligibility
    eligible = get_eligible_cards_for_user(user_row, cards_df)

    if len(eligible) == 0:
        return pd.DataFrame()

    # Step 2: cluster-biased sampling
    sampled = sample_cards_by_cluster(
        user_row,
        eligible,
        total_cards=total_cards
    )

    # Step 3: tier diversity
    final_candidates = enforce_tier_diversity(sampled, eligible)

    return final_candidates
```

```python
def compute_reward_alignment(user_row, card_row):
    user_shares = [
        user_row['share_Groceries'],
        user_row['share_Dining_Food'],
        user_row['share_Transport_Fuel'],
        user_row['share_Retail_Apparel'],
        user_row['share_Digital_Subscriptions'],
        user_row['share_Travel_Lodging'],
        user_row['share_Entertainment_Leisure']
    ]

    card_rewards = [
        card_row['reward_grocery'],
        card_row['reward_dining'],
        card_row['reward_fuel'],
        card_row['reward_retail'],
        card_row['reward_digital'],
        card_row['reward_travel'],
        card_row['reward_other']
    ]

    return sum(u * c for u, c in zip(user_shares, card_rewards))
```

```python
def compute_fee_penalty(user_row, card_row):
    spend = max(user_row['total_spend'], 1.0)
    return card_row['annual_fee'] / spend
```

```python
def compute_capacity_gap(user_row, card_row):
    card_mid_limit = (card_row['credit_limit_min'] + card_row['credit_limit_max']) / 2
    return abs(user_row['avg_credit_limit'] - card_mid_limit) / card_mid_limit
```

```python
def compute_fico_buffer(user_row, card_row):
    return (user_row['fico_score'] - card_row['min_fico']) / 100
```

```python
def compute_cluster_affinity(user_row, card_row):
    return 1.0 if user_row['cluster'] == card_row['cluster_id'] else 0.0
```

```python
def compute_risk_penalty(user_row):
    return (
        2.0 * user_row['pct_critical'] +
        1.0 * user_row['pct_high'] +
        0.5 * user_row['avg_fraud_score']
    )
```

```python
def compute_weak_score(user_row, card_row):
    reward = compute_reward_alignment(user_row, card_row)
    fee_penalty = compute_fee_penalty(user_row, card_row)
    capacity_gap = compute_capacity_gap(user_row, card_row)
    fico_buffer = compute_fico_buffer(user_row, card_row)
    cluster_affinity = compute_cluster_affinity(user_row, card_row)
    risk_penalty = compute_risk_penalty(user_row)

    weak_score = (
        2.5 * reward
        - 1.5 * fee_penalty
        - 1.0 * capacity_gap
        + 0.5 * fico_buffer
        + 0.3 * cluster_affinity
        - 2.0 * risk_penalty
    )

    return weak_score
```

```python
def build_user_interactions(user_row, cards_df):
    candidates = generate_candidate_cards(user_row, cards_df)

    interactions = []

    for _, card_row in candidates.iterrows():
        score = compute_weak_score(user_row, card_row)

        interactions.append({
            'user_id': user_row['User'],
            'card_id': card_row['card_id'],
            'weak_score': score
        })

    return pd.DataFrame(interactions)
```

```python
from tqdm import tqdm

def build_full_interaction_dataset(df4, cards_df):
    """
    Build interaction dataset for all users.
    Each row = (user_id, card_id, weak_score)
    """
    all_interactions = []

    for _, user_row in tqdm(df4.iterrows(), total=len(df4)):
        user_interactions = build_user_interactions(user_row, cards_df)

        if not user_interactions.empty:
            all_interactions.append(user_interactions)

    interaction_df = pd.concat(all_interactions, ignore_index=True)

    return interaction_df
```

```python
def generate_pairs_for_user(user_df, top_k=3, bottom_k=3):
    """
    user_df: interaction rows for ONE user
    returns list of (user_id, pos_card_id, neg_card_id)
    """
    user_df = user_df.sort_values('weak_score', ascending=False)

    top_cards = user_df.head(top_k)
    bottom_cards = user_df.tail(bottom_k)

    pairs = []

    for _, pos in top_cards.iterrows():
        for _, neg in bottom_cards.iterrows():
            if pos['card_id'] != neg['card_id']:
                pairs.append({
                    'user_id': pos['user_id'],
                    'pos_card_id': pos['card_id'],
                    'neg_card_id': neg['card_id']
                })

    return pairs
```

```python
from tqdm import tqdm
import pandas as pd

def build_pairwise_dataset(interaction_df, top_k=3, bottom_k=3):
    all_pairs = []

    for user_id, user_df in tqdm(interaction_df.groupby('user_id')):
        if len(user_df) < (top_k + bottom_k):
            continue  # not enough candidates

        user_pairs = generate_pairs_for_user(
            user_df,
            top_k=top_k,
            bottom_k=bottom_k
        )

        all_pairs.extend(user_pairs)

    pairwise_df = pd.DataFrame(all_pairs)
    return pairwise_df
```

```python
pairwise_df = build_pairwise_dataset(
    interaction_df,
    top_k=3,
    bottom_k=3
)
```
